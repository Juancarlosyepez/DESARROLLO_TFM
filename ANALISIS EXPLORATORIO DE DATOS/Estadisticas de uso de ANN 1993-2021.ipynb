```python
import matplotlib.pyplot as plt

# Datos proporcionados
data = {
    "Ensemble": [2021],
    "IDW-ANN": [2015],
    "GA": [2012],
    "GA-BPNN": [2021],
    "GS-Pred": [2020],
    "RVM": [2015],
    "kNN": [2017, 2020],
    "SVR": [2013, 2019],
    "RF": [2014, 2015, 2016],
    "GP": [2016, 2017, 2018],
    "Neuro-fuzzy": [2007, 2009, 2010, 2012],
    "SVM": [2007, 2011, 2013, 2016, 2018],
    "ANN": [1993, 1996, 1998, 1999, 2002, 2004, 2005, 2006, 2008, 2009, 2010, 2011, 2012, 2015, 2016, 2017, 2018, 2020]
}

# Crear una lista de años únicos
years = list(set(year for years_list in data.values() for year in years_list))
years.sort()

# Contar cuántas veces se usó cada técnica en cada año
techniques_count = {technique: [0] * len(years) for technique in data.keys()}
for technique, technique_years in data.items():
    for year in technique_years:
        index = years.index(year)
        techniques_count[technique][index] += 1

# Configuración de la gráfica
plt.figure(figsize=(12, 6))
plt.bar(years, techniques_count["Ensemble"], label="Ensemble")
plt.bar(years, techniques_count["IDW-ANN"], label="IDW-ANN", bottom=techniques_count["Ensemble"])
plt.bar(years, techniques_count["GA"], label="GA", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"])])
plt.bar(years, techniques_count["GA-BPNN"], label="GA-BPNN", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"], techniques_count["GA"])])
plt.bar(years, techniques_count["GS-Pred"], label="GS-Pred", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"], techniques_count["GA"], techniques_count["GA-BPNN"])])
plt.bar(years, techniques_count["RVM"], label="RVM", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"], techniques_count["GA"], techniques_count["GA-BPNN"], techniques_count["GS-Pred"])])
plt.bar(years, techniques_count["kNN"], label="kNN", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"], techniques_count["GA"], techniques_count["GA-BPNN"], techniques_count["GS-Pred"], techniques_count["RVM"])])
plt.bar(years, techniques_count["SVR"], label="SVR", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"], techniques_count["GA"], techniques_count["GA-BPNN"], techniques_count["GS-Pred"], techniques_count["RVM"], techniques_count["kNN"])])
plt.bar(years, techniques_count["RF"], label="RF", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"], techniques_count["GA"], techniques_count["GA-BPNN"], techniques_count["GS-Pred"], techniques_count["RVM"], techniques_count["kNN"], techniques_count["SVR"])])
plt.bar(years, techniques_count["GP"], label="GP", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"], techniques_count["GA"], techniques_count["GA-BPNN"], techniques_count["GS-Pred"], techniques_count["RVM"], techniques_count["kNN"], techniques_count["SVR"], techniques_count["RF"])])
plt.bar(years, techniques_count["Neuro-fuzzy"], label="Neuro-fuzzy", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"], techniques_count["GA"], techniques_count["GA-BPNN"], techniques_count["GS-Pred"], techniques_count["RVM"], techniques_count["kNN"], techniques_count["SVR"], techniques_count["RF"], techniques_count["GP"])])
plt.bar(years, techniques_count["SVM"], label="SVM", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"], techniques_count["GA"], techniques_count["GA-BPNN"], techniques_count["GS-Pred"], techniques_count["RVM"], techniques_count["kNN"], techniques_count["SVR"], techniques_count["RF"], techniques_count["GP"], techniques_count["Neuro-fuzzy"])])
plt.bar(years, techniques_count["ANN"], label="ANN", bottom=[sum(x) for x in zip(techniques_count["Ensemble"], techniques_count["IDW-ANN"], techniques_count["GA"], techniques_count["GA-BPNN"], techniques_count["GS-Pred"], techniques_count["RVM"], techniques_count["kNN"], techniques_count["SVR"], techniques_count["RF"], techniques_count["GP"], techniques_count["Neuro-fuzzy"], techniques_count["SVM"])])

# Configuración de la gráfica
plt.xlabel("Año")
plt.ylabel("Cantidad de Usos")
plt.title("Técnicas de Machine Learning utilizadas en diferentes yacimientos a lo largo de los años")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))

# Mostrar la gráfica
plt.xticks(years, rotation=90)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
data = {
    "Ensemble": [2021],
    "IDW-ANN": [2015],
    "GA": [2012],
    "GA-BPNN": [2021],
    "GS-Pred": [2020],
    "RVM": [2015],
    "kNN": [2017, 2020],
    "SVR": [2013, 2019],
    "RF": [2014, 2015, 2016],
    "GP": [2016, 2017, 2018],
    "Neuro-fuzzy": [2007, 2009, 2010, 2012],
    "SVM": [2007, 2011, 2013, 2016, 2018],
    "ANN": [1993, 1996, 1998, 1999, 2002, 2004, 2005, 2006, 2008, 2009, 2010, 2011, 2012, 2015, 2016, 2017, 2018, 2020]
}

# Lista de tipos de yacimientos correspondientes a cada técnica
yacimiento_types = {
    "Ensemble": [],
    "IDW-ANN": ["Copper"],
    "GA": ["Iron"],
    "GA-BPNN": ["Copper"],
    "GS-Pred": ["Gold"],
    "RVM": ["Copper"],
    "kNN": ["Gold"],
    "SVR": ["Copper", "Iron"],
    "RF": ["Iron", "Gahnite", "Copper"],
    "GP": ["Copper", "Gold"],
    "Neuro-fuzzy": ["Copper", "Coal"],
    "SVM": ["Platinum", "Copper", "Nickel", "Slate", "Silver", "Gold"],
    "ANN": ["Bauxite", "Gold", "Silver", "Limestone", "Zinc", "Lead", "Iron", "Phosphate", "Copper", "Chromite", "Potash", "Slate", "Gypsum", "Sericite", "Pyrite", "Copper", "Mineral sand"]
}

# Crear una lista de tipos de yacimientos únicos
unique_yacimiento_types = set(yacimiento for yacimientos_list in yacimiento_types.values() for yacimiento in yacimientos_list)

# Calcular la cantidad de usos de "ANN" en cada tipo de yacimiento
ann_usage = [sum(1 for technique, yacimientos_used in yacimiento_types.items() if "ANN" in technique and yacimiento in yacimientos_used) for yacimiento in unique_yacimiento_types]

# Nombres de los tipos de yacimientos
yacimiento_names = list(unique_yacimiento_types)

# Configuración de la gráfica de barras
plt.figure(figsize=(12, 6))

# Crear la gráfica de barras
plt.bar(yacimiento_names, ann_usage, color='b', alpha=0.7)

# Configuración de la gráfica
plt.xlabel("Tipo de Yacimiento")
plt.ylabel("Cantidad de Usos de ANN")
plt.title("Frecuencia de uso de ANN en diferentes tipos de yacimientos")

# Ajustar la escala vertical para mostrar correctamente las frecuencias
plt.ylim(0, max(ann_usage) + 1)

# Rotar etiquetas en el eje x para mejorar la legibilidad
plt.xticks(rotation=90)

# Mostrar la gráfica de barras
plt.tight_layout()
plt.show()
```
