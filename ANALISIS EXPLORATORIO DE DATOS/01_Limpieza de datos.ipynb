## Verificamos caracteres especiales en los datos
se verifico caracteres especiales del tipo mayor que (>) y menor que (<) los cuales eliminaremos

```python
import re

# Cargar el DataFrame df desde el archivo CSV
#df = pd.read_csv('ruta_del_archivo.csv', sep=',')

# Define la expresión regular para buscar caracteres especiales (puedes modificarla según tus necesidades)
pattern = r'[!@#$%^&*(),?":{}|<>]'

# Aplica la expresión regular a cada valor del DataFrame y verifica si hay coincidencias
mask = df.applymap(lambda x: bool(re.search(pattern, str(x))))

# Filtra las filas que tienen caracteres especiales en alguna de las celdas
rows_with_special_chars = df[mask.any(axis=1)]

# Muestra las filas que tienen caracteres especiales en alguna de las celdas
print(rows_with_special_chars)
```

```python
# Define la expresión regular para buscar caracteres especiales (en este caso, buscará > y < seguidos de un número)
pattern = r'[!@#$%^&*(),?":{}|<>]'

# Verificar qué valores tienen caracteres especiales en las columnas numéricas
for col in df.columns[2:]:
    # Aplica la expresión regular a cada valor de la columna
    mask = df[col].astype(str).str.contains(pattern)
    if mask.any():
        print(f"Valores con caracteres especiales en la columna '{col}':")
        print(df[col][mask])

```

## Elimanos los caracteres especiales identificados

```python
# Definir una función para eliminar el carácter especial y convertir el valor a número
def eliminar_caracter_especial(val):
    if pd.notna(val):
        match = re.match(r'[!@#$%^&*(),?":{}|<>] ?(\d+(\.\d+)?)', val)
        if match:
            return float(match.group(1))
    return val

# Aplicar la función a las columnas con caracteres especiales
df['Ca'] = df['Ca'].apply(eliminar_caracter_especial)
df['Zr'] = df['Zr'].apply(eliminar_caracter_especial)

```

## Verficando duplicados de muestras columna from - to

```python
# Creamos nuevas columnas llamadas 'From_antes' y 'To_antes' que contendrán los valores de 'From' y 'To' desplazados una fila hacia arriba
df['From_antes'] = df['From'].shift(1)
df['To_antes'] = df['To'].shift(1)

# Verificamos si los valores de 'From' y 'To' son iguales a los valores 'From_antes' y 'To_antes' en registros consecutivos
df['Es_Duplicado'] = ((df['From'] == df['From_antes']) & (df['To'] == df['To_antes']))

# Filtramos el DataFrame original para obtener solo aquellos registros donde tanto 'From' como 'To' se repiten seguido del anterior
df_duplicados_seguidos = df[df['Es_Duplicado']]

# Eliminamos las columnas 'From_antes', 'To_antes' y 'Es_Duplicado' si no las necesitamos en el resultado final
#df_duplicados_seguidos = df_duplicados_seguidos.drop(columns=['From_antes', 'To_antes', 'Es_Duplicado'])
df_duplicados_seguidos.to_csv('duplicados_from_to.csv', index=False, encoding='utf-8')
df_duplicados_seguidos
```

## Promediado de duplicados y eliminacion de los duplicados

```python
# Creamos nuevas columnas llamadas 'From_antes' y 'To_antes' que contendrán los valores de 'From' y 'To' desplazados una fila hacia arriba
df['From_antes'] = df['From'].shift(1)
df['To_antes'] = df['To'].shift(1)

# Verificamos si los valores de 'From' y 'To' son iguales a los valores 'From_antes' y 'To_antes' en registros consecutivos
df['Es_Duplicado'] = ((df['From'] == df['From_antes']) & (df['To'] == df['To_antes']))

# Iteramos a través de los registros duplicados en el DataFrame original
for index, row in df[df['Es_Duplicado']].iterrows():
    # Obtenemos el índice del registro anterior al duplicado
    indice_anterior = index - 1

    # Obtenemos el registro anterior al duplicado
    registro_anterior = df.loc[indice_anterior]

    # Calculamos el promedio de las leyes geoquímicas en los registros duplicados
    promedio_leyes = df.loc[[indice_anterior, index]].iloc[:, 3:].mean(skipna=True)

    # Actualizamos el registro anterior con el promedio
    df.iloc[indice_anterior, 3:] = promedio_leyes.values

# Identificamos los registros duplicados y los eliminamos manteniendo solo el primero de ellos
df_limpio = df.drop_duplicates(subset=['DDH','From', 'To'], keep='first')

# Eliminamos las columnas 'From_antes', 'To_antes' y 'Es_Duplicado' ya que no las necesitamos en el resultado final
df_limpio = df_limpio.drop(columns=['From_antes', 'To_antes', 'Es_Duplicado'])


# Exportamos el DataFrame limpio con los duplicados promediados a un archivo CSV
df_limpio.to_csv('dataset_limpio.csv', index=False, encoding='utf-8')

df_limpio.head(41)
```
